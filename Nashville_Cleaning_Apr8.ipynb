### Back to cleaning df2

```python
# convert 'time' column to datetime type
df2['time'] = pd.to_datetime(df2['time'])
```

```python
# extract time component
df2['time'] = df2['time'].dt.time
```

```python
df2
```

```python
# convert time column to datetime format with a placeholder date
df2['time'] = pd.to_datetime('2022-01-01 ' + df2['time'].astype(str))

# extract hour component from datetime format
df2['hour'] = df2['time'].dt.hour

# bin the hour values into desired time buckets
bins = [0, 2, 10, 24]
labels = ['10pm-6am', '6am-2pm', '2pm-10pm']
df2['time_bucket'] = pd.cut(df2['hour'], bins=bins, labels=labels, include_lowest=True)
```

```python
df2['time_bucket'] = df2['time_bucket'].replace(
    {'6am-2pm': 'Morning', 
     '2pm-10pm': 'Afternoon',
    '10pm-6am': 'Night'})
```

```python
df3 = df2.groupby(['date', 'precinct', 'time_bucket']).size().reset_index(name='count')
```
